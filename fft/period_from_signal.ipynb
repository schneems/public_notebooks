## Visualizing FFT

This shows how to use numpy FFT. The first example uses a signal we generate, after that I'm using real X and Y coordinates 

```python
## This shows how to demonstrate a FFT on a sine wave

import matplotlib.pyplot as plt
import plotly.plotly as py
import numpy as np

Fs = 150.0;  # sampling rate
Ts = 1.0/Fs; # sampling interval
t = np.arange(0,1,Ts) # time vector

ff = 5;   # frequency of the signal
y = np.sin(2*np.pi*ff*t)

n = len(y) # length of the signal
k = np.arange(n)
T = n/Fs
frq = k/T # two sides frequency range
frq = frq[range(n/2)] # one side frequency range

Y = np.fft.fft(y)/n # fft computing and normalization
Y = Y[range(n/2)]

fig, ax = plt.subplots(2, 1)
ax[0].plot(t,y)
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Amplitude')
ax[1].plot(frq,abs(Y),'r') # plotting the spectrum
ax[1].set_xlabel('Freq (Hz)')
ax[1].set_ylabel('|Y(freq)|')

plt.show()

False
```

## Time lagged correlation

From a class discussion:

So a cross-correlation of two signals a(n) and b(n) is computed by

a∗b=c(l)=∑na(n)b(n+l)

Think of it as the dot product of the two sequences offset by l. If a(n) and b(n) are sinusoids with a phase offset of m, then c(l) has a maximum at l=m.  For this application, the important point is that max(c(l)) gives the maximum possible dot product between the signals after correcting for a phase offset.

with no distance x ~= cos(2π/T * n)

```python
from math import *

wave_forms_for_period = {}
period_range = [29, 30, 31]
for period in period_range:
    wave_forms_for_period[period] = []
    for i in range(len(x_array)):
        value  = cos(2 * pi / period * i)
        wave_forms_for_period[period].append(value)

time_array = np.arange(len(x_array))

print "Different periods give us different waveforms"
plt.plot(time_array[0:50], wave_forms_for_period[29][0:50], '-r')
plt.plot(time_array[0:50], wave_forms_for_period[30][0:50], '-b')
plt.plot(time_array[0:50], wave_forms_for_period[31][0:50], '-g')
plt.show()
```

## Add waveforms to signal

If we add these waveforms for different periods to our signal we can see where they join together. We have to "correlate" the waveforms, we can't simply multiply them (or add them) because they may be offset. scipi's signal.correlate fixes this for us out of the box

```python
print "We add these waveforms to our signal"

x_signal = np.array(x_array)

period29 = signal.correlate(x_signal, wave_forms_for_period[29])
period30 = signal.correlate(x_signal, wave_forms_for_period[30])
period31 = signal.correlate(x_signal, wave_forms_for_period[31])


time_array = np.arange(len(period30))

plt.plot(time_array, period29, '-r')
plt.plot(time_array, period30, '-b')
plt.plot(time_array, period31, '-g')
plt.show()
```

```python
print "We find the higest value and that is the true period"

max_peak = 0
actual_period = None
for period in period_range:
    peak = np.max(signal.correlate(x_signal, wave_forms_for_period[period]))
    if peak > max_peak:
        max_peak      = peak
        actual_period = period

print "Expected period is 30, actual value is:", actual_period
```
